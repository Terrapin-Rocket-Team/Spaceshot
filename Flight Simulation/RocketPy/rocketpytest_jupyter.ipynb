### Import

```python
from rocketpy import Environment, SolidMotor, Rocket, Flight
import datetime
```

### Define Environment
#### Higgs Farm (weather 24 hours from time of sim)

```python
# Location
env = Environment(latitude=39.0785417, longitude=-75.8763458, elevation=30) # Higgs Farm Elevation
# Time
tomorrow = datetime.date.today() + datetime.timedelta(days=1)
env.set_date((tomorrow.year, tomorrow.month, tomorrow.day, 12)) # Hour given in UTC time
# Atmospheric Model
env.set_atmospheric_model(type="Forecast", file="GFS")
# Print Weather
print(env.info())
```
